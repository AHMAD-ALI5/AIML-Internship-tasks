```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
def health_chatbot(user_query):
    prompt = f"""
    You are a helpful and friendly medical assistant.

    Rules:
    - Provide general health information only
    - Do NOT give diagnosis
    - Do NOT prescribe medication
    - Keep answers simple and easy to understand
    - Always suggest consulting a doctor for serious issues

    User Question: {user_query}
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
def safety_filter(query):
    blocked_words = ["dose", "prescribe", "treatment plan", "exact medicine"]

    for word in blocked_words:
        if word in query.lower():
            return False
    return True
```

```python
def safe_chatbot(query):
    if not safety_filter(query):
        return "⚠️ Please consult a doctor for medical advice."

    return health_chatbot(query)
```

```python
print(safe_chatbot("What causes a sore throat?"))
print(safe_chatbot("Is paracetamol safe for children?"))
```

```python
while True:
    user_input = input("Ask a health question (type exit to stop): ")

    if user_input.lower() == "exit":
        break

    response = safe_chatbot(user_input)
    print("Bot:", response)
```
